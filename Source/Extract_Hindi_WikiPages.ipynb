#### 1. Importing Libraries

```python
import sys
import requests
import bs4
import json
import re
```

#### 2. Extracting WikiPages for all Hindi Labelled Monuments

```python
with open('../Data/Wikidata_JSON/hi_monument_dump.json') as f:
    hi_monument_list = json.load(f)
```

```python
articles = {}
for monument in hi_monument_list:
    res = requests.get('https://hi.wikipedia.org/wiki/' + monument['labels']['hi']['value'])
    if res.status_code != 404:
        print(monument['labels']['hi']['value'])
        wiki = bs4.BeautifulSoup(res.text,"lxml")
        elems = wiki.select('p')
        article = ''
        for i in range(len(elems)):
            article += elems[i].getText()
    
        article = article.replace('\n', '')
        article = article.rsplit('।',1)[0]+'।'
        articles[monument['id']] = article
```

```python
print(len(articles))
with open('../Data/Hindi_Wikipages/hi_articles.json', 'w') as f:
    json.dump(articles, f)
```
